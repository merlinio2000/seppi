$$ \hat{w} = \sum_{i=1}^{n} m_{i} * B^{\hat{e} - i} $$

$ m_{i} $ = i-te Stelle der Mantisse <br>
$ \hat{e} $ = Exponent e im Dezimalsystem <br>

$ x_{min} = B^{e_{min}-1} \\$

### Fixpunktiteration

$$ F(x) = x \\ $$

Iteration der Form: <br>
$x_{n+1} = F(x_n) \\$

$ \text{Nherungslsung Fixpunkt: } \bar{x} = x_n \\  $
$ \text{Anziehender Fixpunkt: } |F'(\bar{x})| \le 1 \\ $
$ \text{Abstossender Fixpunkt: } |F'(\bar{x})| > 1 \\ $


#### Banachscher Fixpunktsatz

$$ F: [a, b] \mapsto [a, b] \enspace \\ $$ 
$$ \& \enspace \exists \alpha : |F(x)-F(y)| \le \alpha|x-y| \text{ fr x, y } \in [a,b]\\ $$

Lipschitzwert: $ \alpha = max |F'(x_0)| : x_0 \in [a, b] \\$
wobei $ 0 < \alpha < 1$

***dann gilt*** (f ist Libschitz-stetig)

* F hat genau einen Fixpunkt $ \bar{x} \in [a, b] $
* Iteration konvergiert gegen $ \bar{x} \forall x_0 \in [a, b] $
* a-priori
$ |x_n - \bar{x}| \le \frac{\alpha^n}{1-\alpha} |x_1 - x_0| \\ $
* a-posteriori
$ |x_n - \bar{x}| \le \frac{\alpha}{1-\alpha} |x_n - x_{n-1}| \\ $

## Fehlerrechnung und Aufwandabschtzung

### Normen

$$
||x|| \ge 0 \; \& \; ||x|| = 0 \lrarr x = 0 \\
||\lambda x|| = |\lambda| * ||x|| \\
||x + y|| \le ||x|| + ||y|| 
$$

Vektor und Matrixnormen sind "kompatibel": $ ||Ax||_i \le ||A||_i * ||x||_i$

**Fr worst-case Rechnung:** $\infty$-Norm

#### Vektornormen

- 1-Norm, Summennorm: $||x||_1 = \displaystyle\sum^n_{i=1} |x_i|$
- 2-Norm, euklidische Norm: $||x||_2 = \sqrt{\displaystyle\sum^n_{i=1} x_i^2}$
- $\infty$-Norm, Maximumnorm: $||x||_{\infty} = max(|x_i|)$

#### Matrixnormen

- 1-Norm, Spaltensummennorm: $||A||_1 = \displaystyle\max_{j=1,...,n}(\displaystyle\sum^n_{i=1} |A_{ij}|)$
- 2-Norm, Spektralnorm: $||A||_2 = \sqrt{\rho(A^TA)}$
- $\infty$-Norm, Zeilensummennorm: $||A||_{\infty} = \displaystyle\max_{i=1,...,n}(\displaystyle\sum^n_{j=1} |A_{ij}|)$

$\rho(...)$ steht hierbei fr den Spektralradius


### Fehlerabschtzung
$$Ax=b \rarr \text{ Fehlerbehaftet: } A\tilde{x}=\tilde{b}$$

- absolut: $||x - \tilde{x}|| \le ||A^{-1}|| * ||b-\tilde{b}||$
- relativ: $\frac{||x-\tilde{x}||}{||x||} \le ||A|| * ||A^{-1}|| * \frac{||b-\tilde{b}||}{||b||}$ falls $||b|| \ne 0$

#### Konditionszahl (von A bzgl. verwendeter Norm)
$$
cond(A) = ||A|| * ||A^{-1}||
$$
grosse Konditionszahl -> fehler verstrkt / schlecht konditionierte Matrix


#### Fehlerbehaftete Matrix

$$
\text{Falls: } cond(A) * \frac{||A-\tilde{A}||}{||A||} < 1 \\
\\
\text{dann gilt: } \\
\frac{||x-\tilde{x}||}{||x||} \le \frac{cond(A)}{1-(cond(A)*\frac{||A-\tilde{A}||}{||A||})}
* \left( \frac{||A-\tilde{A}||}{||A||} + \frac{||b-\tilde{b}||}{||b||} \right)
$$


## Iterative Verfahren

$x^{(k)} \text{steht hierbei fr x nach der k-ten Iteration,\red{ NICHT POTENZ}}$

### Jacobi-Verfahren (Gesamtschritt-Verfahren)

$A$ zerlegen zu $A = L + D + R$ (unter-,sie selbst, oberhalb der Diagonale)
$$
Dx^{(k+1)} = -(L+R)x^{(k)} + b \; \text{ bzw.} \\
x^{(k+1)} = -D^{-1}(L+R)x^{(k)} + D^{-1}b 
$$

wobei sich die untere Variante nur lohnt, wenn die inverse von $D$ effizient berechnet werden kann

Als Algorithmus: <br>

$
x_i^{(k+1)} = \frac{1}{a_{ii}}\left( b_i - 
    \displaystyle\sum^{n}_{j=1,j\ne i} a_{ij} x_j^{(k)}
 \right) \quad i=1,...,n
$


### Gauss-Seidel-Verfahren (Einzelschrittverfahren)

$A$ zerlegen zu $A = L + D + R$ (unter-,sie selbst, oberhalb der Diagonale)
$$
(D + L)x^{(k+1)} = -Rx^{(k)} + b \; \text{ bzw.} \\
x^{(k+1)} = -(D + L)^{-1}Rx^{(k)} + (D + L)^{-1}b 
$$

Als Algorithmus:  <br>

$
x_i^{(k+1)} = \frac{1}{a_{ii}} \left(
    b_i
    - \displaystyle\sum^{i-1}_{j=1}{a_{ij} x_j^{k+1}}
    - \displaystyle\sum^{n}_{j=i+1}{a_{ij} x_j^{(k)}}
\right) \quad i=1,...,n
$



## Konvergenz

Gegeben: Matrix $B \in \R^{n \times n}$, Vektor $\vec c \in \R^n$ und Fixpunkiteration
$$
x^{(n+1)} = Bx^{(n)} + \vec c =: F(x^{(n)})
$$

- Jacobi: 
$ B = -D^{-1}(L + R) $
- Gauss:
$ B = -(D + L)^{-1}R $

#### anziehender / abstossender Fixpunkt

Fixpunkt $\bar x = B \bar x + \vec c = F(\bar x)$ ist
- anziehend: $||B|| < 1$
- abstossend: $||B|| > 1$

**Diagonaldominanz** ist ebenfalls ein hinreichendes Kriterium (es gibt allerdings Matrizen die konvergieren und nicht diagonaldominant sind) <br>
Dieses Kriterium erfllt falls: <br>
- $\forall\; i = 1,...,n: |a_{ii}| > \sum^n_{j=1, j\ne i}|a_{ij}|$
ODER
- $\forall\; j = 1,...,n: |a_{ii}| > \sum^n_{i=1, i\ne j}|a_{ij}|$

sprich alle Diagonalelemente betragsmssig grsser alls die zugehrige Spalte <br>
ODER alle grsser als die zugehrige Reihe (KEIN MISCHEN)

#### Abschtzung (bei anziehendem Fixpunkt)

Die Formeln sind analog zum Banachschen Fixpunktsatz mit $\alpha = ||B||$

- a-priori: 
$ \displaystyle ||x^{(n)} - \bar x|| \le \frac{||B||^n}{1-||B||} ||x^{(1)} - x^{(0)}|| $

- a-posteriori:
$ \displaystyle ||x^{(n)} - \bar x|| \le \frac{||B||}{1-||B||} ||x^{(n)} - x^{(n-1)}|| $


Zum a-priori **Abschtzen der Anzahl bentigten Schritte** kann die Formel wie folgt umgeformt werden:

$$
\displaystyle
n \ge \frac{ln \left( \frac{tol * (1 - \alpha)}{||x^{(1)} - x^{(0)}||} \right)}
        {ln(\alpha)}
$$

## Komplexe Zahlen

$$ i^2 = -1 $$

Damit lassen sich gleichungen wie $x^2 + 1 = 0$ mit $ x = i $ lsen


Eine komplexe Zahl $z$ besteht aus einem real-Teil $x$ und einem imaginr-Teil $y$ <br>
Dies lsst sich als Punkt $P = (x, y)$ auf einer Zahlenebene darstellen mit: <br>
$$ z = x + iy \\
Re(z) = x \quad Im(z) = y $$

Daraus folgt auch die definition der komplexen Zahlen $\cnums =\{z|z=x+iy, \; x,y \in \R \}$

### betrag

Entspricht der Lnge des "Pfeil" zu $P=(x,y)$ in der Zahlenebene

$$|z| = \sqrt{x^2 + y^2}$$

### komplex konjugiert

Entspricht der Spiegelung von $z$ an der $x$-Achse
$$ z^* = x - iy \\
z * z^* = x^2 + y^2 = |z|^2$$

### Formen

#### Normalform (kartesische Form)

$$z = x + iy$$

#### Trigonometrische Form

$$z = r(cos(\varphi) + i * sin(\varphi))$$

dabei gilt: <br>
- $x = r * cos(\varphi)$
- $y = r * sin(\varphi)$

#### Exponentialform

$$z = re^{i\varphi}$$

### Grundrechenarten

#### Addition
$$ z_1 + z_2 = (x_1 + x_2) + i(y_1 + y_2)$$

#### Subtraktion
$$ z_1 - z_2 = (x_1 - x_2) + i(y_1 - y_2)$$

#### Multiplikation
$$ z_1 * z_2 =  (x_1 x_2 - y_1 y_2 ) + i( x_1 y_2 + x_2 y_1 ) $$

#### Division
$$ \frac{z_1}{z_2}\begin{align} 
                    &= \frac{z_1 * z_2^*}{z_2 * z_2^*} \\
                    &= \frac{(x_1 x_2 + y_1 y_2) + i (y_1 x_2 - x_1 y_2)}
                            {x_2^2 + y_2^2} \\
                    \end{align} $$

### Grundgesetze (NUR * und +)

#### Kommutativ
$$ z_1 + z_2 = z_2 + z_1 \quad | \quad z_1 * z_2 = z_2 * z_2 $$

#### Assoziaziv
$$ 
z_1 + (z_2 + z_3) = (z_1 + z_2) + z_3 \\
z_1 * (z_2 * z_3) = (z_1 * z_2) * z_3 
$$

#### Distributiv
$$
z_1 * (z_2 + z_3) = z_1 * z_2 + z_1 * z_3
$$


### Potenzieren und Radizieren

Gegeben $z$ Komplex und $n \in \N$

$$
z^n = (re^{i\varphi})^n = r^n e^{in\varphi} = r^n * (cos(n\varphi) + i * sin(n\varphi))
$$

$$

und 2 komplexen Lsungen $z_2 = 2i \, \& \, z_3 = -2i$ wobei $z_2^* = z_3$ und umgekehrt

$$

## Eigenwerte und Eigenvektoren

Sei $A \in \R^{n \times n}, \, \lambda \in \cnums$ heisst dann **Eigenwert von $A$** <br>
wenn es einen Vektor $\vec{x} \in \cnums^n \backslash \{0\} (d.h. \vec{x} \ne \vec{0})$

$$
A\vec{x} = \lambda \vec{x}
$$

$\vec{x}$ heisst dann **Eigenvektor von $A$**

Ist $\lambda_i$ ein Eigenwert von $A$, so ist $\frac{1}{\lambda_i}$ Eigenwert der inversen Matrix $A^{-1}$

<br>

### Eigenwerte bestimmen
Durch umformen: 
$A\vec x - \lambda \vec x = \vec 0 \iff (A -\lambda I_n) \vec x = \vec 0$ 

wobei dieses Gleichungssystem nur eine Lsung $\vec x \ne \vec 0$ hat wenn 
$det(A - \lambda I_n) = 0 $

### Bestimmung durch charakteristisches Polynom / Spur
Sei $A \in \R^{n \times n}, \lambda \in \cnums$, dann gilt: <br>
$$
\lambda \text{ ist Eigenwert von } A \iff det(A-\lambda I_n) = 0
$$

Die Abbildung $p$
$$
p: \lambda \mapsto det(A- \lambda I_n)
$$
ist Polynom von Grad $n$ und wird **charakteristisches Polynom von $A$** genannt.

Die Eigenwerte von $A$ sind also alle Nullstellen von $p$. $A$ hat somit genau $n$ Eigenwerte, von denen manche mehrfach vorkommen knnen

#### Determinante aus Eigenwerten
$det(A) = \lambda_1 * \lambda_2 * ... * \lambda_n$

#### Spur
$tr(A) = a_{11} + a_{22} + ... + a_{nn} = \lambda_1 + \lambda_2 + ... + \lambda_n $


### Algebraische Vielfachheit / Spektrum

Die Vielfachheit mit der $\lambda$ als Nullstelle des charaktertistischen Polynoms von $A$ auftritt heisst **algebraische Vielfachheit von $\lambda$**

Das Spektrum $\sigma(A)$ ist die Menge aller Eigenwerte von $A$


### Eigenschaften von Eigenvektoren

Bei zwei Eigenvektoren $x, y$ zum selben Eigenwert $\lambda \in \cnums$
ist auch die Summe $x + y$ sowie jedes Vielfache von $x$ ein Eigenvektor zum Eigenwert $\lambda$

$$
\begin{aligned}
 A(x+y) &= Ax + Ay = \lambda x + \lambda y = \lambda(x + y) \\
 A(\mu x) &= \mu A x = \mu \lambda x = \lambda \mu x
\end{aligned}
$$

#### Eigenraum von $\lambda$

Die Eigenvektoren zum gleichen Eigenwert $\lambda$ bilden einen Untervektorraum von $\cnums^n$ wenn man den Nullvektor (welcher nicht als Eigenvektor zhlt) hinzunimmt.

Es gibt also unendlich viele Eigenvektoren zu $\lambda$ (Linearkombinationen der Basisvektoren des Eigenraums)

Bei der Bestimmung der Eigenvektoren treten $n - r$ freie Parameter auf. <br>
Dabei ist $r = rg(A - \lambda I_n) < n$ Rang der Koeffizientenmatrix. <br>
Die Bestimmung der Eigenvektoren entspricht also der Bestimmung der Basis des Eigenraums

Der Eigenraum ist die Lsungsmenge von:
$$
(A - \lambda I_n)x = 0
$$
Welches nur eine Lsung $x \ne 0$ hat wenn $r = rg(A - \lambda I_n) < n$


Die Dimension des Eigenraums von $\lambda$ ($= r$) wird **geometrische Vielfachheit von \lambda** genannt. Dies Entspricht der Anzahl linear unabhniger Eigenvektoren in der Basis des Eigenraums. <br>
Die geometrische Vielfachheit ist stets $\le$ algebraische Vielfachheit.

- sind alle $n$ Eigenwerte verschieden, so hat jeder Eigentwert genau einen linear unabhngigen Eigenvektor
- Tritt ein Eigenwert $k$-Fach auf (d.h. algebraische Vielfachheit $k$) so gehren dazu mindestens 1, hchstens $k$ linear unabhngige Eigenvektoren auf
- die verschiedenen Eigenvektoren zu einem Eigenwert sind immer linear unabhngig


**Eigenvektoren zu konjugiert komplexen Eigenwerten sind ebenfalls zueinander konjugiert komplex**


### Numerische Berechnung von EW/EV

#### hnliche Matrizen / Diagonalisierbarkeit

mit Matrizen $A, B \in \R^{n \times n}$ und **regulre Matrix** $T$ mit<br>
$$
B = T^{-1} A T
$$
so heissen $B$ und $A$ zueinander hnliche Matrizen. "$B$ geht aus $A$ durch hnlichkeitstransformation hervor" <br>

Spezialfall $B = D$ eine Diagonal/Dreiecksmatrix also 
$$D = T^{-1} A T \; \rarr \; A \text{ diagonalisierbar}$$

1. $A$ und $B$ haben die selben Eigenwerte ink. deren algebraische Vielfachheit
2. Ist $\vec x$ ein EV zum EW $\lambda$ von $B$, dann ist $T\vec x$ ein EV zum EW $\lambda$ von A
3. Im Spezialfall $B = D$ ("A diagonaliserbar") sind die $n$ Diagonalelemente von $D$ die Eigenwerte von $A$ und die $n$ **normierten und linear unabhngigen** Eigenvektoren stehen in den Spalten von $T$


#### Berechnung mit QR-Verfahren

Idee:
QR-Zerleung von $A_0 = Q_1 R_1$ mit auflsen nach $R_1 = Q_1^T A$ <br>
Die Beidseitige Multiplikation mit $Q_1$ ergibt dann:

$$
A_1 = R_1 Q_1 = Q_1^T A_0 Q_1 \quad ...\text{repeat}
$$
wobei gilt: $A_1$ hnelt $A$ und das Verfahren fr $A_\infty$ gegen eine obere Dreiecksmatrix konvergiert. <br>
In dieser oberen Dreiecksmatrix stehen nurnoch die reelen Eigenwerte auf der Diagonale respektive komplexe Eigenwerte in $2 \times 2$ Blcken an der Diagonale anliegend (Diagonale des Blocks = Diagonalre der $2 \times 2$ sub-Matrix)

#### Algorithmus
$
\rarr A \in \R^{n \times n} \\
A_0 := A \\
P_0 := I_n \\
for \; i = 0, 1, 2,... \\
\quad Q_i, R_i := qr(A_i) \; \text{\# Berechne QR Zerlegung von } A_i \\
\quad A_{i+1} := R_i * Q_i \\
\quad P_{i+1} := P_i * Q_i \\
$

Bemerkungen:
1. Falls alle Eigenwerte **betragsmssig** verschieden sind hat $A$ genau $n$ reele Eigenwerte und es treten keine $2 \times 2$ Blcke auf. Gennant "perfekte obere Dreiecksmatrix"
2. Nur EWs werden berechnets, EVs noch separat
3. Ist $A_0$ symmetrisch ($A_0^T = A_0$) und alle Eigenwerte **betragsmssig** verschieden so stehen EV in den Spalten der $P_k (\equiv T)$





### Vektoriteration

#### Spektralradius (siehe Spektralnorm)

entspricht dem Betrag des **betragsmssig grssten** Eigenwert

$$\rho (A) = max(|\lambda| \; | \; \lambda \text{ ist Eigenwert von }A \in R^{n \times n})$$


#### von-Mises-Iteration (engl. Power Method)

liefert den EW des Spektralradius sowie zugehrigen Eigenvektor. Konvergiert fr
fast jeden Startvektor $v^{(0)} \in \cnums^n$ mit Lnge 1 ($\equiv$ normiert)

$$
\begin{aligned}
v^{(k+1)} &= \frac{A v^{(k)}}{||A v^{(k)}||_2} \\
\\
\lambda^{(k+1)} &= \frac{(v^{(k)})^T * A * v^{(k)}}{(v^{(k)})^T v^{(k)}}
\end{aligned}
$$






## Repetition

### Determinanten

- Diagonalmatrix: $\displaystyle\prod_{i=0}^{n} a_{ii} \quad $ "Produkt der Diagonalelemente"

- $2 \times 2$
$$
A = 
\begin{pmatrix}
a & b \\
c & d
\end{pmatrix}
\; \rarr \;  det(A) = ac - bd 
$$


### Linear unabhngig

$n$ Vektoren $e_1,...,e_n \in \cnums^n$ sind linear unabhngig falls <br>
$$
\mu_1 e_1 + ... + \mu_n e_n \\
\text{nur fr } \mu_1 = ... = \mu_n = 0 \text{ erfllt werden kann}
$$

### Rang einer Matrix

Der Rang einer Matrix $A$ entspricht der Anzahl linear unabhngiger Zeilenvektoren von $A$.<br>
Lsst sich z.B. mit Gauss machen (Zeilen $\ne 0$)

### Vektorraum

- Dimension eines Vektorraums ist die Anzahl der Basisvektoren
- Die Basis ist eine Menge linear unabhngiger Vektoren, jeder Vektor im Vektorraum lsst sich durch eine Linearkombination dieser darstellen.
